```python
# REMOVING SENIOR, ADVANCED AND DIRECTOR LEVEL JOBS AS FROM THE INTRODUCTORY COURSE WE ARE NOT GOING TO HAVE SENIOR LEVEL POSITIONS 
df = df[~df['JOB TITLE'].str.contains('Senior|senior|Sr|sr|Advanced|Director', regex=True)]
df.reset_index(drop=True, inplace=True) 
```

```python
# DEFINING A FUNCTION TO REMOVE BRACKETS 
def listToStringWithoutBrackets(list1):
    return str(list1).replace('[','').replace(']','')
```

```python
# REMOVING BRACKETS FROM JOB DESCRIPTION 
df['JOB DECSRIPTION'] = df['JOB DECSRIPTION'].apply(lambda x : listToStringWithoutBrackets(x))
df['JOB DECSRIPTION'].head(1)
```

```python
# REMOVING BRACKETS FROM JOB TITLES
df['JOB TITLE'] = df['JOB TITLE'].apply(lambda x : listToStringWithoutBrackets(x))
df['JOB TITLE'].head(1)
```

```python
# DEFINING A FUNCTION TO REMOVE PUNCTUATIONS
def remove_punct(text):
    punct = "".join([i for i in text if i not in string.punctuation])
    return punct
```

```python
# cleaning of job description
def clean1(text):#converting to lowercase
    lower_case = text.lower()
    return(lower_case)
def clean2(lowercase):#removing punctuations
    clean_lines=lowercase.replace("\\n", "")
    clean_lines = re.sub(r"[,@\?\\.$%_\\\']", "", clean_lines)
    clean_lines=BeautifulSoup(clean_lines, 'lxml').get_text()
    return(clean_lines)
# Function for Splitting sentences into words using tokens
def clean3(clean2):#tokenizing
        tokenizer = RegexpTokenizer(r'\w+')         
        b = tokenizer.tokenize(clean2)
        return(b)
stop_words1=set(stopwords.words('english'))#remove stop words
def clean4(clean3):
        clean4= [word for word in clean3 if word not in stop_words1]#removing words not in english
        return(clean4)
```

### 5.Feature Extraction and Model testing


Now that we have the skills for each job description,lets find out if the skills extracted from the manually created dictionary really represents the nature of jobs(i.e high level job and medium level job). 
This is done by applying logistic regression on the extracted skills to predict the class of the job posting.   
This is done to justify that the manually created dictionary of skills infact represents the skills for both high level and medium level jobs

#### Model-1 for Feature Extraction

```python
#Bag of words (from sklearn.feature_extraction.text import CountVectorizer)
def bag(X):
    vectorizer1 = CountVectorizer(max_features =5000)
    X1 = vectorizer1.fit_transform(X)
    X1_n = normalize(X1)#normalize the obtained feature vectors(features are skills)
    z1=pd.DataFrame(X1.toarray(), columns=vectorizer1.get_feature_names())
    return(X1_n,z1)
```

#### Model-2 for Feature Extraction

```python
#TF-IDF(from sklearn.feature_extraction.text import TfidfVectorizer
def tfidf(X):
    vectorizer2 = TfidfVectorizer(use_idf=True, smooth_idf=True, norm=None, max_features=5000)
    X2 = vectorizer2.fit_transform(X)
    X2_n = normalize(X2)
    z2=pd.DataFrame(X2.toarray(), columns=vectorizer2.get_feature_names())
    return(X2_n,z2)
```

#### Logistic Regression for predicting multiclass job titles to find accuracy of the model1 and model2(for best features)

```python
#Splitting into training and testing sets
def allmodels(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=200, test_size=0.3)
    
    #Applying logistic regression
    model=LogisticRegression()
    model.fit(X_train, y_train)
    predictions=model.predict(X_test)
    acc_logr=round(accuracy_score(y_test,predictions)*100,2)
    print('Accuracy for bag of words model is {}%'.format(acc_logr))
    return(acc_logr)
```
